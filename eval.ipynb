```python
import pandas as pd
import numpy as np

# Assuming your dataset is loaded into a pandas DataFrame called 'df'
# Let's say the labels are in a column named 'label'

# Define the number of samples you want from each category
num_samples = 70

# Separate the data into the categories
no_fit = df.iloc[:3143]
potential_fit = df.iloc[3143:4699]
good_fit = df.iloc[4699:]

# Randomly sample 70 data from each category
no_fit_sample = no_fit.sample(n=num_samples, random_state=42)
potential_fit_sample = potential_fit.sample(n=num_samples, random_state=42)
good_fit_sample = good_fit.sample(n=num_samples, random_state=42)

# Combine the samples into a new dataframe
sampled_df = pd.concat([no_fit_sample, potential_fit_sample, good_fit_sample])

# Shuffle the combined dataframe to ensure randomness across categories
sampled_df = sampled_df.sample(frac=1, random_state=42).reset_index(drop=True)
```
